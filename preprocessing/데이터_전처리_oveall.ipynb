```python
import geopandas as gpd
from geopy.distance import geodesic
import pandas as pd

# Shapefile   
shp_file_path = r'C:\Users\1102w\Desktop\\\ \_.shp'
gdf = gpd.read_file(shp_file_path, encoding='utf-8')

#  WGS 84 
gdf = gdf.to_crs(epsg=4326)

#    
grid_centroids = gdf.geometry.centroid
grid_geometry = gdf.geometry  #   geometry ( )

#     (: df_candidate    )
# df_candidate : df_candidate = pd.DataFrame({'': [...], '': [...]})

# df_candidate       
candidate_nodes = list(zip(df_candidate[''], df_candidate['']))

#   
distance_matrix = []
for centroid in grid_centroids:
    grid_coord = (centroid.y, centroid.x)  #   (, ) 
    grid_distances = []
    for node_coord in candidate_nodes:
        distance = geodesic(grid_coord, node_coord).meters  #   ( )
        grid_distances.append(distance)
    distance_matrix.append(grid_distances)

#   
distance_df = pd.DataFrame(distance_matrix, columns=[f"Node_{i+1}" for i in range(len(candidate_nodes))])

# `gdf` `distance_df` : geometry  gid 
distance_df['geometry'] = grid_geometry.values  #  geometry 
distance_df = distance_df.merge(gdf[['gid', 'geometry']], on='geometry', how='left')  # geometry  

#  gid  geometry  
distance_df.set_index('gid', inplace=True)

#   ( 5  )
sampled_indices = distance_df.sample(5).index
for idx in sampled_indices:
    original_geom = gdf.loc[gdf['gid'] == idx, 'geometry'].values[0]
    matched_geom = distance_df.loc[idx, 'geometry']
    print(f"GID: {idx}")
    print(f"Original Geometry: {original_geom}")
    print(f"Matched Geometry: {matched_geom}")
    print("Match:", original_geom.equals(matched_geom), "\n")

#    
all_matched = all(
    gdf.loc[gdf['gid'] == idx, 'geometry'].values[0].equals(distance_df.loc[idx, 'geometry'])
    for idx in distance_df.index
)
print("All geometries match:", all_matched)
```

```python
import geopandas as gpd
from shapely.wkt import loads

# geometry  
print("Before conversion:")
print(df_h_variable['geometry'].head())

# geometry  WKT   
if isinstance(df_h_variable['geometry'].iloc[0], str):
    df_h_variable['geometry'] = df_h_variable['geometry'].apply(loads)

# GeoDataFrame 
df_h_variable = gpd.GeoDataFrame(df_h_variable, geometry='geometry')

#  
if df_h_variable.crs is None:
    df_h_variable.set_crs(epsg=4326, inplace=True)

# 
print("After conversion:")
print(df_h_variable.head())
```

```python
# Spatial Join 
joined = gpd.sjoin(df_h_variable, gdf[['geometry', 'Day_Total']], how='left', predicate='intersects')

#     
joined = joined[~joined.index.duplicated(keep='first')]

#   df_h_variable 
df_h_variable_merged = df_h_variable.copy()
df_h_variable_merged['Day_Total'] = joined['Day_Total']

#   
print(df_h_variable_merged.head())

#   
total_rows = len(df_h_variable_merged)
matched_rows = df_h_variable_merged['Day_Total'].notna().sum()
unmatched_rows = total_rows - matched_rows

#  
print(f"Matched geometries: {matched_rows} / {total_rows}")
print(f"Unmatched geometries: {unmatched_rows} / {total_rows}")
```

```python
# (NaN) 0 
df_h_variable_merged['Day_Total'] = df_h_variable_merged['Day_Total'].fillna(0)

#   
df_h_variable_merged
```

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler 
scaler = MinMaxScaler()

# val Day_Total  
df_h_variable_merged[['val_normalized', 'day_total_normalized']] = scaler.fit_transform(
    df_h_variable_merged[['val', 'Day_Total']]
)

#    
df_h_variable_merged['result'] = (
    df_h_variable_merged['val_normalized'] * 0.7 + 
    df_h_variable_merged['day_total_normalized'] * 0.3
)

# result  val  
df_h_variable_merged['val'] = df_h_variable_merged['result']

#  
print(df_h_variable_merged[['val', 'Day_Total', 'result']].head())
```
